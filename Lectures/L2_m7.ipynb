## 🐼 Module 7: Pandas

Pandas is used for working with tabular data (like spreadsheets).


```python
import pandas as pd

data = {
    "name": ["Alice", "Bob", "Cathy"],
    "score": [88, 92, 95]
}

df = pd.DataFrame(data)
print(df)

print("Average score:", df["score"].mean())
```
