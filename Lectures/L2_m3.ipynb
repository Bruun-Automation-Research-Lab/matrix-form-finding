## 📦 Module 3: Conda, Environments, and Packages

Conda helps manage Python versions and packages. Use environments to isolate projects.

### 🔧 Setup
```bash
conda create -n myenv python=3.10
conda activate myenv
conda install numpy pandas matplotlib scipy
```

### 🌐 Useful Commands
```bash
conda list
conda env list
conda env export > environment.yml
conda env create -f environment.yml
```

### 📝 Practice
1. Try creating a new environment on your own machine.
2. Install `jupyterlab` in that environment and launch it.
