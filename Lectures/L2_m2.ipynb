## 🧱 Module 2: Object-Oriented Programming (OOP)

Use classes to bundle data and behavior.


```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

r = Rectangle(4, 5)
print("Area:", r.area())
```
