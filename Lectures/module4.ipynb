## 🔢 Module 4: NumPy

NumPy allows efficient numerical operations on arrays.


```python
import numpy as np

a = np.array([1, 2, 3])
print("Array:", a)
print("Mean:", np.mean(a))
print("Reshape:", a.reshape((3, 1)))
```
