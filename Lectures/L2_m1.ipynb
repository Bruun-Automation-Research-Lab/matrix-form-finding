## 📘 Module 1: Python Basics

This module introduces variables, data types, control flow, and functions.


```python
# Numbers, strings, booleans
x = 42
name = "Alice"
is_student = True

print(f"{name} is a student: {is_student}, age {x}")
```

```python
# Conditional logic
if x > 30:
    print("x is large")
else:
    print("x is small")
```

```python
# Loops
for i in range(5):
    print(f"Iteration {i}")
```

```python
# Functions
def double(n):
    return n * 2

print(double(4))
```
